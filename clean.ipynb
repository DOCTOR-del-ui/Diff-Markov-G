```python
import os
```

```python
addname = "markovpluswf1i"
```

```python
for idx in range(1,5):
    for window in [48, 96]:
        data_path_ground = os.path.join("OUTPUT", f"cmapsst{addname}{idx}_{window}", "samples", f"cmapsst{addname}{idx}_ground_truth_{window}_train.npy")
        data_path_norm = os.path.join("OUTPUT", f"cmapsst{addname}{idx}_{window}", "samples", f"cmapsst{addname}{idx}_norm_truth_{window}_train.npy")
        
        if os.path.exists(data_path_ground):
            os.remove(data_path_ground)
        if os.path.exists(data_path_norm):
            if addname != "":
                os.remove(data_path_norm)
                
        check_dir = os.path.join("Checkpoints", f"Checkpoints_cmapsst{addname}{idx}_{window}")
        if "mini" in addname:
            filiter_name = "checkpoint-50.pt"
        elif "proplus" in addname:
            filiter_name = "checkpoint-300.pt"
        elif "plus" in addname:
            filiter_name = "checkpoint-150.pt"
        else:
            filiter_name = "checkpoint-100.pt"
            
        if os.path.exists(check_dir):
            for filename in os.listdir(check_dir):
                if filename != filiter_name:
                    pathcurr = os.path.join(check_dir, filename)
                    os.remove(pathcurr)
```
